# Testing the network class with normally distributed data
In this notebook we present how to work with the network class in order to classify data.  
The model behind this is taken from [C.F. Higham, D.J. Higham; Deep Learning: An Introduction for Applied Mathematicians, arXiv (2018)]

Here, we consider the simple case of two dimensional input data. The training data consists of two sets of normally distributed points in $\mathbb{R}^2$ with two different expectation values.
Let us first import some necessary modules.

```python
import numpy as np
import network_class as nc
from matplotlib import pyplot as plt
from functools import partial
from itertools import permutations
```

## Generation of training data
We will consider normally distributed data and use the following function in order to generate data.

```python
def genNormal(mu, sigma, dim, num):
    """Generate num np.arrays of length dim, sampled from a normal distribution,
    with expectation mu and variance sigma."""
    return [np.random.normal(mu, sigma, dim) for i in range(num)]

dim_training_data = 2
num_data = 100
mu1, mu2 = 0.1, 0.35
sigma1, sigma2 = 0.1, 0.1
input1 = genNormal(mu1, sigma1, dim_training_data, num_data)
label1 = [True for i in range(len(input1))]
input2 = genNormal(mu2, sigma2, dim_training_data, num_data)
label2 = [False for i in range(len(input2))]
training_data = input1 + input2
labels = label1 + label2
```

### Visualization of training data
We have chosen the parameters of the normal distribution such that there is an overlap of the training data.
The different labelling of the data is visualized by red dots and blue stars.

```python
plt.figure()
xdata1 = np.array([x[0] for x in input1])
ydata1 = np.array([x[1] for x in input1])
xdata2 = np.array([x[0] for x in input2])
ydata2 = np.array([x[1] for x in input2])
plt.plot(xdata1, ydata1, 'ro', xdata2, ydata2, 'b+')
plt.legend(["$\mu_1=$" + str(mu1) + ", $\sigma_1=$" + str(sigma1) , "$\mu_2=$" + str(mu2) + ", $\sigma_2=$" + str(sigma2)])
plt.title("Normally distributed test data")
plt.show()
```

### Random initialization of the network
For the classification task we consider a deep network with two-dimensional input and output layers and two hidden layers of sizes $3$ and $4$ resp.
In order to apply the neural network module we randomly initialize the network's weights and set the bias to zero. For faster convergence we scale the network's weights closer to zero by the factor "scaling".

```python
dim_layers = [dim_training_data, 3, 4, 2]
num_layers = len(dim_layers)
weight_matrices = []
biases = []
scaling = 0.15
for i in range(num_layers - 1):
    weight_matrices.append(scaling * np.random.rand(dim_layers[i+1], dim_layers[i]))
    biases.append(np.zeros(dim_layers[i+1]))
network = nc.NN(weight_matrices, biases)
for matrix in network.weights:
    print(matrix)
```

## Optimize network with training data
Now, we are ready to apply the method optimNetw from the NN class to the freshly instantiated object "network".

```python
network.optimNetw(training_data, labels, nc.sigmoid, nc.target2d)
for matrix in network.weights:
    print(matrix)
```

### Test optimized network with unlabelled data
After having optimized the network's weights, we can test its ability to classify normally distributed data regarding their expectation value. Simply evaluate the netwtork function netw_out on a given data point and check, whether the result is closer (in euclidean distance) to the value of training data with label 1 or label 2.
Here, we choose test data from a distribution with expectation $\mu$ slightly differing from $\mu_2$, so it should be classified to data set 2.

```python
num_test_data = 100
shift = 0.05
mu = mu2 - shift
test_data = genNormal(mu, sigma2, dim_training_data, num_test_data)
output = []
success = []
for data in test_data:
    out = network.netw_out(data, nc.sigmoid)
    output.append(out)
    dist1 = np.linalg.norm(out-np.array([1, 0]))
    dist2 = np.linalg.norm(out-np.array([0, 1]))
    if dist2 < dist1:
        success.append(True)
    else:
        success.append(False)  
    
success_rate = sum(success)/num_test_data
plt.figure()
x_data_success = np.array([value[0][0] for value in zip(test_data, success) if value[1]])
y_data_success = np.array([value[0][1] for value in zip(test_data, success) if value[1]])
x_data_fail = np.array([value[0][0] for value in zip(test_data, success) if not value[1]])
y_data_fail = np.array([value[0][1] for value in zip(test_data, success) if not value[1]])
plt.plot(x_data_success, y_data_success, 'g+', x_data_fail, y_data_fail, 'ko', mew=1.5, ms=8)
plt.plot( xdata1, ydata1, 'ro', xdata2, ydata2, 'b+', alpha=0.5)
plt.legend(["Success", "Fail", "Training set 1" , "Training set 2"])
plt.title("Classification of data with $\mu = \mu_2$ -" + str(shift) + ". Success rate is " + str(success_rate))
plt.show()
    
```

### Determining the classification boundary
Let us next determine the boundary curve which is obtained by the classification scheme. More precisely, we classify points on a grid in $[0,1]\times[0,1]$ and determine the curve along which classification swaps from label 1 to label 2.

```python
nx, ny = (200, 200)
x = np.linspace(0, 0.6, nx)
y = np.linspace(0, 0.6, ny)
xv, yv = np.meshgrid(x, y)
curve = []
for k in range(ny):
    l = 0
    label1 = True
    while label1 and l < nx:
        out = network.netw_out(np.array([xv[l,k],yv[l,k]]), nc.sigmoid)
        dist1 = np.linalg.norm(out-np.array([1, 0]))
        dist2 = np.linalg.norm(out-np.array([0, 1]))
        if dist1 < dist2:
            l += 1
        else:
            label1 = False
            curve.append(0.5*(np.array([xv[l,k],yv[l,k]]) + np.array([xv[l-1,k],yv[l-1,k]])))
# And plotting
plt.figure()
x_data_curve = np.array([x[0] for x in curve])
y_datacurve = np.array([x[1] for x in curve])

plt.plot(x_data_curve, y_datacurve, 'k')
plt.plot(xdata1, ydata1, 'ro', xdata2, ydata2, 'b+', alpha=0.5)
plt.legend(["Curve of separation", "Training set 1" , "Training set 2"])
plt.title("Curve of separation of classification scheme")
plt.show()
        
        
```

In the above figure, points below the black curve are classified as label 1, so belonging to a normal distribution with expectation and variance $\mu_1$ and $\sigma_1$ resp. Points above the curve are classified as label 2 and hence belonging to a normal distribution with expectation and variance $\mu_2$ and $\sigma_2$ .
